```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
probs = pd.read_csv('../output/preds/mean_[2v+4r+2ir+i+x]_p025.csv', index_col='fname').values
# probs = pd.read_csv('../output/preds/mean_[2v+4r+2ir+i+x].csv', header=None).values

labels = np.argmax(probs, axis=1)
max_probs = np.max(probs, axis=1)

LABELS = ['yes', 'no', 'up', 'down', 'left', 'right', 'on', 'off', 'stop', 'go', 'silence', 'unknown']
ID2NAME = {i: name for i, name in enumerate(LABELS)}
NAME2ID = {name: i for i, name in ID2NAME.items()}

for l in range(12):
    plt.figure(figsize=(12, 6))
    counts, bins, patches = plt.hist(max_probs[labels == l], bins=10)
    print(counts)
    plt.title(ID2NAME[l])
    plt.show()
    
```
